```python
import numpy as np
import skimage.io as io

# Read the image file
image = io.imread("Desktop/Neuron_in_tissue_culture.jpg")

# Create separate Red and Green images based on the filtering threshold of 100
red_image = np.zeros_like(image)
green_image = np.zeros_like(image)

red_indices = image[:, :, 0] > 100
green_indices = image[:, :, 1] > 100

red_image[red_indices, 0] = image[red_indices, 0]
green_image[green_indices, 1] = image[green_indices, 1]

# Save the Red and Green images
io.imsave("Desktop/Red_image.jpg", red_image)
io.imsave("Desktop/Green_image.jpg", green_image)
```

```python
# Calculate the number of red pixels at 100 threshold
num_red_pixels = np.sum(red_indices)

# Calculate the number of green pixels at 100 threshold
num_green_pixels = np.sum(green_indices)

print(f"Number of red pixels: {num_red_pixels}")
print(f"Number of green pixels: {num_green_pixels}")
```

```python
# Create separate Red and Green images based on the new filtering threshold of 150
image = io.imread("Neuron_in_tissue_culture.jpg")

red_image_threshold_150 = np.zeros_like(image)
green_image_threshold_150 = np.zeros_like(image)

red_indices_threshold_150 = image[:, :, 0] > 150
green_indices_threshold_150 = image[:, :, 1] > 150

red_image_threshold_150[red_indices_threshold_150, 0] = image[red_indices_threshold_150, 0]
green_image_threshold_150[green_indices_threshold_150, 1] = image[green_indices_threshold_150, 1]

# Calculate the number of red pixels with the new threshold
num_red_pixels_threshold_150 = np.sum(red_indices_threshold_150)

# Calculate the number of green pixels with the new threshold
num_green_pixels_threshold_150 = np.sum(green_indices_threshold_150)

print(f"Number of red pixels with threshold 150: {num_red_pixels_threshold_150}")
print(f"Number of green pixels with threshold 150: {num_green_pixels_threshold_150}")
```

```python
# Save the Red and Green images at 150 threshold
io.imsave("Desktop/Red_image.jpg", red_image)
io.imsave("Desktop/Green_image.jpg", green_image)
```
